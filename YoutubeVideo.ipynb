```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Generating sample data
np.random.seed(0)
X = np.linspace(-1, 1, 100).reshape(-1, 1)  #set the input
y = 2 * X + np.random.normal(0, 0.2, size=X.shape)

# Creating a sequential model
model = Sequential([
    Dense(10, input_dim=1, activation='relu'),  # Hidden layer with 10 neurons and ReLU activation
    Dense(1)                                     # Output layer with 1 neuron (linear activation by default)
])

# Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Training the model
history = model.fit(X, y, epochs=50, batch_size=10, verbose=0)

# Plotting the training loss
plt.plot(history.history['loss'])
plt.title('Model Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()

# Making predictions
X_test = np.array([[0.5], [-0.5]])
predictions = model.predict(X_test)

# Plotting the original data and predictions
plt.scatter(X, y, label='Original data')
plt.plot(X_test, predictions, color='red', label='Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Original Data vs. Predictions')
plt.legend()
plt.show()
```

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from keras.applications import MobileNetV2
from keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
from keras.preprocessing import image

# Load the pre-trained MobileNetV2 model
model = MobileNetV2(weights='imagenet')

# Function to preprocess the image
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    return preprocess_input(img_array)

# Function to predict the class of the image
def predict_image_class(img_path):
    processed_img = preprocess_image(img_path)
    predictions = model.predict(processed_img)
    decoded_predictions = decode_predictions(predictions, top=3)[0]
    return decoded_predictions

# Replace 'YOUR_IMAGE_URL' with your own image URL
img_url = 'https://imgs.search.brave.com/xlew05k8FHpnrefFG09TLNNzraI8J2yzP8eqyuA6DM0/rs:fit:500:0:0/g:ce/aHR0cHM6Ly9tZWRp/YS5pc3RvY2twaG90/by5jb20vaWQvMTM4/NTIxNzk2OS9waG90/by9naW5nZXItY2F0/LXdhbGtzLmpwZz9z/PTYxMng2MTImdz0w/Jms9MjAmYz1tQmhG/ekRBMmRwMjNkVjRq/cTZGb29hU3pHMnRt/Q1pNS1A2VFY1NmhN/VklJPQ'

# Downloading the sample image
!wget $img_url -O sample_image.jpg

# Predicting the class of the sample image
predictions = predict_image_class('sample_image.jpg')

# Displaying the predictions
for i, (imagenet_id, label, score) in enumerate(predictions):
    print("{}. {}: {:.2f}%".format(i+1, label, score*100))

# Displaying the sample image
img = image.load_img('sample_image.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
```
